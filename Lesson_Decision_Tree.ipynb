```python
import numpy as np
import warnings
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import linear_model

from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
breast = load_breast_cancer()
X, Y = breast.data, breast.target
X_train, X_test, y_train, y_test = train_test_split (X, Y,
                                                    train_size=0.75,
                                                    random_state=42)

"""**DecisionTreeClassifier** """

tree_classifier = DecisionTreeClassifier(random_state=1)
tree_classifier.fit(X_train, y_train)

Y_preds = tree_classifier.predict(X_test)
print('Test Accuracy : %.3f'%tree_classifier.score(X_test, y_test))
print('Train Accuracy : %.3f'%tree_classifier.score(X_train, y_train))

acc_train = []
acc_test = []
for i in range(1,11):
  tree_classifier = DecisionTreeClassifier(random_state=1, max_depth=i)
  tree_classifier.fit(X_train, y_train)
  acc_train.append(tree_classifier.score(X_train, y_train))
  acc_test.append(tree_classifier.score(X_test, y_test))

plt.plot(acc_test)
plt.plot(acc_train)
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('max_depth')
plt.legend(['test', 'train'], loc='upper left')
plt.show()

```

```python
"""**RandomForestClassifier**"""

RFC_acc_train = []
RFC_acc_test = []
for i in range(2,21):
  RFC = RandomForestClassifier(n_estimators=i, random_state=1)
  RFC.fit(X_train, y_train)
  RFC_acc_train.append(RFC.score(X_train, y_train))
  RFC_acc_test.append(RFC.score(X_test, y_test))

plt.plot(RFC_acc_test)
plt.plot(RFC_acc_train)
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('n_estimators')
plt.legend(['test', 'train'], loc='upper left')
plt.show()

```

```python
"""**2 задание**"""

np.random.seed(0)
X = np.arange(1.0, 10.0, 0.1).reshape(-1, 1)
Y = np.array([max(0, (1/x) + np.random.normal(scale = 0.05)) for x in X])

plt.scatter(X, Y)
plt.show()

X_train, X_test, y_train, y_test = train_test_split (X, Y,
                                                    train_size=0.75,
                                                    random_state=42)

"""**лин регрессия**"""

reg = linear_model.LinearRegression()
reg.fit(X_train, y_train)
print('R2 = ', reg.score(X_train, y_train))
print('R2 = ', reg.score(X_test, y_test))

"""**дерево регрессии**"""
print()
regTree = DecisionTreeRegressor(random_state=0)
regTree.fit(X_train, y_train)
print('R2 = ', regTree.score(X_train, y_train))
print('R2 = ', regTree.score(X_test, y_test))



```
